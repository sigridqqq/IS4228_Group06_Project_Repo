## 1. Load review data with language label 

```python
import pandas as pd 
from langdetect import detect
import pickle

def save_obj(obj, name):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)

# Read object saved from language detect python script
# The resultant dataframe should be one with ever review labelled with its language type

check = load_obj('/home/q/qshichao/BT4222/original_review_language_detect')
check
```

```python
# We only keep the english reviews in this project

only_english = check[check['language'] == 'en']
only_english
```

## 2. Create function to remove punctuation, some non-English characters still remaining in English reviews, expand text to remove short forms or abbreviations, and normalize text

```python
import contractions
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
sb = SnowballStemmer(language = 'english')
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

def remove_punct(token):
 return [word for word in token if word.isalpha()]

def remove_nonenglish_char(text):
  cleaned = []
  sent = word_tokenize(text)
  sent = remove_punct(sent)
  for words in sent:
    if words.isalpha():
      cleaned.append(words)
  return ' '.join(cleaned)

def expand_text(text):
  expanded_words = []
  for word in text.split():
    expanded_words.append(contractions.fix(word))
  expanded_text = ' '.join(expanded_words)
  return expanded_text

def normalize_text(text):
  stop_words = set(stopwords.words('english'))
  sent = word_tokenize(text)
  sent = remove_punct(sent)
  normalized_sent = []
  for words in sent:
    processed_word = lemmatizer.lemmatize(words)
    if processed_word.lower() not in stop_words:
      normalized_sent.append(processed_word)
  return ' '.join(normalized_sent)
```

```python
only_english['cleaned_text'] = only_english['text'].apply(remove_nonenglish_char)
only_english
```

```python
import numpy as np

only_english['cleaned_text'].replace('', np.nan, inplace=True)
only_english.dropna(subset=['cleaned_text'], inplace=True)
only_english
```

```python
nltk.download('omw-1.4')
only_english['cleaned_text'] = only_english['cleaned_text'].apply(expand_text)
only_english['cleaned_text'] = only_english['cleaned_text'].apply(normalize_text)
only_english
```

## 4. We doublecheck that there are no other non-English characters, especially Chinese, Japanese or Korean, still remain in the review text

```python
import re
indexlist = []
regex = re.compile('[^\u0020-\u024F]')
# regex.sub('', train.iloc[896188,-1])
for x in range(len(train)):
    if re.search(u'[\u3040-\u30ff\u3400-\u4dbf\u4e00-\u9fff\uf900-\ufaff\uff66-\uff9f]', train.iloc[x,-1]):
        indexlist.append(x)
        train.iloc[x,-1] = regex.sub('', train.iloc[x,-1])
```

```python
for x in range(len(val)):
    if re.search(u'[\u3040-\u30ff\u3400-\u4dbf\u4e00-\u9fff\uf900-\ufaff\uff66-\uff9f]', val.iloc[x,-1]):
        val.iloc[x,-1] = regex.sub('', val.iloc[x,-1])

for x in range(len(test)):
    if re.search(u'[\u3040-\u30ff\u3400-\u4dbf\u4e00-\u9fff\uf900-\ufaff\uff66-\uff9f]', test.iloc[x,-1]):
        test.iloc[x,-1] = regex.sub('', test.iloc[x,-1])
```
