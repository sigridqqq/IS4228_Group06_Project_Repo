```python
import contractions
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
sb = SnowballStemmer(language = 'english')
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
def remove_punct(token):
 return [word for word in token if word.isalpha()]

def remove_nonenglish_char(text):
  cleaned = []
  sent = word_tokenize(text)
  sent = remove_punct(sent)
  for words in sent:
    if words.isalpha():
      cleaned.append(words)
  return ' '.join(cleaned)

def expand_text(text):
  expanded_words = []
  for word in text.split():
    expanded_words.append(contractions.fix(word))
  expanded_text = ' '.join(expanded_words)
  return expanded_text

def normalize_text(text):
  stop_words = set(stopwords.words('english'))
  sent = word_tokenize(text)
  sent = remove_punct(sent)
  normalized_sent = []
  for words in sent:
    processed_word = lemmatizer.lemmatize(words)
    if processed_word.lower() not in stop_words:
      normalized_sent.append(processed_word)
  # sent = [sb.stem(words_sent) for words_sent in sent]
  # sent = [lemmatizer.lemmatize(words_sent) for words_sent in sent]
  # sent = [w for w in sent if not w.lower() in stop_words]
  return ' '.join(normalized_sent)
```

```python
yelp_t_12 = yelp_t.iloc[: , 1:]
yelp_t_12
```

```python
yelp_t_12['cleaned_text'] = yelp_t_12['text'].apply(remove_nonenglish_char)
yelp_t_12
```

```python
yelp_t_12['cleaned_text'].replace('', np.nan, inplace=True)
yelp_t_12.dropna(subset=['cleaned_text'], inplace=True)
yelp_t_12
```

```python
yelp_t_12['cleaned_text'] = yelp_t_12['cleaned_text'].apply(expand_text)
yelp_t_12['cleaned_text'] = yelp_t_12['cleaned_text'].apply(normalize_text)
```

```python
yelp_t_12_dropped = yelp_t_12.drop(['text'], axis=1)
yelp_t_12_dropped.to_csv('yelp_train_processed.csv',index=False)
```
